```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(
    data="AFDET_dataset.yaml",
    epochs=100,
    imgsz=640,
    hsv_h=0.0,         # Hue augmentation (probability/factor)
    hsv_s=0.0,         # Saturation augmentation
    hsv_v=0.0,         # Value (brightness) augmentation
    translate=0.0,     # Image translation
    scale=0.0,         # Image scale
    shear=0.0,         # Image shear
    perspective=0.0,   # Image perspective
    flipud=0.0,        # Image flip up-down (probability)
    fliplr=0.0,        # Image flip left-right (probability)
    mosaic=0.0,        # Mosaic augmentation (probability)
    mixup=0.0,         # MixUp augmentation (probability)
    erasing=0.0,       # Erasing augmentation (probability)
    auto_augment=None, # Disable auto augment
    )
```
